```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
import warnings


```

```python
warnings.filterwarnings('ignore')
```

```python
tickers =[ "META",'AAPL','NVDA','AMZN','GOOG','MSFT']
stock_data = yf.download(tickers, start="2020-01-01", end="2024-01-01")
```

```python
df = stock_data
```

```python
df.head()
```

```python
adj_close = df['Adj Close']
```

```python
adj_close.describe
```

market capitalization

```python
open_volume = df[['Open','Volume']]
```

```python
mkt = pd.DataFrame()

mkt ['AAPL_mkt'] = open_volume['Open','AAPL'] * open_volume['Volume',"AAPL"]
mkt ['AMZN_mkt'] = open_volume['Open','AMZN'] * open_volume['Volume',"AMZN"]
mkt ['GOOG_mkt'] = open_volume['Open','GOOG'] * open_volume['Volume',"GOOG"]
mkt ['META_mkt'] = open_volume['Open','META'] * open_volume['Volume',"META"]
mkt ['MSFT_mkt'] = open_volume['Open','MSFT'] * open_volume['Volume',"MSFT"]
mkt ['NVDA_mkt'] = open_volume['Open','NVDA'] * open_volume['Volume',"NVDA"]
```

```python
plt.rcParams['figure.figsize'] = (20, 5)
```

```python
mkt['AAPL_mkt'].plot(label ='APPLE')
mkt['AMZN_mkt'].plot(label ='APPLE')
```
